## Pandas Basic

### Series 

* One-dimensional labeled arrays capable of holding data of any type (integer, string, float, python objects, etc.)
* Axis labels are collectively called index

```python
#try it out yourself with other data types!

pd.Series(data = [1,2,3,'a'], index = ['a', 'b', 'c', 'd'])
```

### DataFrames


* DataFrames are multi-dimensional tables made up of a collection of Series
* Columns give the names of what the data is describing
* Index helps us easily access data in a certain row

* There are two main ways to create DataFrames:
    1. list of lists or dictionary (ordered collection of observations representing a row in the table)
    2. lists keyed by column name (collection of measurements representing the columns of the table)
 

```python
#list of lists example

columns = ['Fruit', 'Store', 'Color', 'In Season']

data = [
    ['Strawberries', 'Vons', 'Red', True],
    ['Oranges', 'Trader Joes', 'Orange', False],
    ['Raspberries', 'Ralphs', 'Red', False],
    ['Mangoes', 'Whole Foods', 'Yellow', True],
]

fruits = pd.DataFrame(data, columns=columns)
fruits
```

```python
#lists keyed by column name example

column_dict = {
    'Fruit': ['Strawberries', 'Oranges', 'Raspberries', 'Mangoes'],
    'Store': ['Vons', 'Trader Joes', 'Ralphs', 'Whole Foods'],
    'Color': ['Red', 'Orange', 'Red', 'Yellow'],
    'In Season': [True, False, False, True]
}

fruits = pd.DataFrame(column_dict)
fruits
```

### Accessing Values in a DataFrame 
* Can use either .loc or [ ] 
    * df.loc[idx] selects the row indexed by the value idx
    * df[col] selects the column labeled by the value col

```python
#.loc example
fruits.loc[1:2]
```

```python
#[] example
fruits['Store']
```

## Basic Exploratory Data Analysis
* matplotlib: comprehensive library for creating static, animated, and interactive visualizations in Python
* seaborn: Python data visualization library based on matplotlib (a bit more complex than matplotlib)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
blockbuster_df = pd.read_csv('blockbusters.csv')
blockbuster_df.head(3) 

#are the types the way we want them to be?
#blockbuster_df.dtypes
```

```python
blockbuster_df['worldwide_gross'] = blockbuster_df['worldwide_gross'].str.strip('$')
blockbuster_df['worldwide_gross']  = blockbuster_df['worldwide_gross'].str.replace(',', '')
blockbuster_df['worldwide_gross'] = blockbuster_df['worldwide_gross'].astype(float)
```

### Line Graph
* What if we wanted to visualize how the length of movies have changed over the years?

```python
#Have to group the years
year_grouped = blockbuster_df.groupby('year').mean()
```

```python
movies_plot = year_grouped['length'].plot()
plt.ylabel('length')
```

### Correlation Matrix
* What if we wanted to see how different variables correlate with each other? (EX: Is there a correlation between worldwide gross and imdb rating?)

```python
# Compute the correlation matrix
corr = blockbuster_df.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 11))

# Draw the correlation matrix 
sns.heatmap(corr, cmap="Blues", square=True, linewidths=3, annot = True)
```

### Horizontal Bar Chart
* What if we wanted to know how much money each top movie in 2018 made? 

```python
recent_movies = blockbuster_df[blockbuster_df['year'] == 2018]
#recent_movies.head(3)
```

```python
plt.barh(recent_movies['title'],recent_movies['worldwide_gross'])
plt.xticks(rotation = 90)
```
